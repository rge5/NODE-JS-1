Q2.  Understanding the Node.js Module System
    Questions
     What is the Node.js module system
      How are modules created and loaded in Node.js
      Create a simple custom module math.js that exports a function to add two numbers, and demonstrate
      loading and using this module in a separate script.

      -> In Node.js, a module is a standalone block of code that encapsulates related functionality. Modules help in
organizing code, making it easier to maintain, test, and reuse.

1. How Modules Are Created

CommonJS (CJS) Module Creation

```
// math.js
function add(a, b) {
  return a + b;
}

module.exports = { add };
```
ES Module (ESM) Creation

```
// math.js
export function add(a, b) {
  return a + b;
}
```

2. How Modules Are Loaded

CommonJS Loading (require)
```
const math = require('./math');
console.log(math.add(2, 3));
```
ES Module Loading (import)
```
import { add } from './math.js';
console.log(add(2, 3));
```


Q3: Write a Node.js script to read the contents of a file synchronously and log it to the console.

->
```
// Import the built-in file system module
const fs = require('fs');

// Specify the file path
const filePath = 'example.txt';

try {
  // Read file synchronously
  const data = fs.readFileSync(filePath, 'utf8');

  // Log file contents
  console.log("File Content:\n", data);
} catch (error) {
  // Handle errors (e.g., file not found)
  console.error("Error reading file:", error.message);
}
```

Q4.Write a Node.js script to rename a file named oldFile.txt to newFile.txt.

```
const fs = require('fs');

fs.rename('oldFile.txt', 'newFile.txt', (err) => {
  if (err) {
    console.error("Error renaming file:", err.message);
    return;
  }
  console.log("File renamed successfully!");
});
```

Q5. Working with OS Module
Questions


*   What is the OS module in Node.js
*   Write a script that logs the system's total memory and free memory using    the os module.

->
The os module is used to retrieve system-level details like CPU info, memory, platform, uptime, and user environment.

```
const os = require('os');

// Get total and free memory (in bytes)
const totalMemory = os.totalmem();
const freeMemory = os.freemem();

// Convert bytes to MB for readability
const toMB = (bytes) => (bytes / (1024 * 1024)).toFixed(2);

console.log("Total Memory:", toMB(totalMemory), "MB");
console.log("Free Memory:", toMB(freeMemory), "MB");
```
